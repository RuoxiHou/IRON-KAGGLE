```python
import pandas as pd
```

```python
data = pd.read_csv('sales.csv')
```

```python
data['date'] = pd.to_datetime(data['date'])
data['year'] = pd.to_datetime(data['date']).dt.year
data['month'] = pd.to_datetime(data['date']).dt.month
data
data.groupby(['open','state_holiday','school_holiday', 'day_of_week'])['index'].agg(['count'])
```

```python
data['year'].unique()
data['state_holiday'] = data['state_holiday'].apply(lambda x: 1 if x in ['a', 'b', 'c'] else 0)
data['state_holiday'].unique()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
# Assuming 'data' is your DataFrame containing the relevant columns

# Extracting predictor and target variables
x = data[['day_of_week', 'nb_customers_on_day', 'open', 'promotion', 'state_holiday', 'school_holiday','month']]
y = data[['sales']]

# Initializing and fitting the model
model = LinearRegression()
result = model.fit(x, y)

# Printing intercept and coefficients
print("Intercept:", result.intercept_)
print("Coefficients:", result.coef_)

# Calculating the regression line
regression_line = result.predict(x)

# Plotting
#plt.plot(y, regression_line, c='orange')
#plt.scatter(y, x)  # You might want to specify which column of 'x' to scatter against 'y'
#plt.xlabel('Sales')
#plt.ylabel('Predicted Sales')
#plt.show()

y_pred = model.predict(x)
r_squared = r2_score(y, y_pred)
print(r_squared)

# Your response here. 
import statsmodels.api as sm
# Agregar una constante al predictor (intercepto)
X = sm.add_constant(data[['day_of_week', 'nb_customers_on_day', 'open', 'promotion', 'state_holiday', 'school_holiday','month']])

# Ajustar el modelo de regresiÃ³n lineal mÃºltiple
model = sm.OLS(data['sales'], X).fit()

# Obtener un resumen completo
summary = model.summary()
print(summary)
# p-valores inferiores a 0.05, Adj. R-squared cercano a 1. Modelo robusto
```
